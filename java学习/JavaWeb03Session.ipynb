# Session(重点)
## 什么是Session：
    服务器会给每一个用户创建一个Session对象
    一个session独占一个浏览器，只要浏览器没有关闭就存在
    用户登录之后整个网站都可以登录——>保存用户信息等等
## Session与Cookie的区别
* Cookie是把用户的数据写给用户的浏览器，浏览器保存（可以保存多个）
* Session把用户的数据写到用户独占的Session中，服务器端保存（保存重要信息，减少浪费）
* Session对象有服务器创建
## 使用场景
* 保存一个用户的信息
* 购物车信息
* 在整个网站中经常会使用的数据
## 使用Session
```java
package com.kuang.servlet;

import com.kuang.pojo.Person;

import javax.servlet.ServletException;
import javax.servlet.http.*;
import java.io.IOException;

public class SessionDemo01 extends HttpServlet {
    @Override
    protected void doGet(HttpServletRequest req, HttpServletResponse resp) throws ServletException, IOException {
        //解决乱码问题
        req.setCharacterEncoding("UTF-8");
        resp.setCharacterEncoding("UTF-8");
        //返回响应页面
        resp.setContentType("text/html;charset=utf-8");//将text转成html界面，然后设置格式为utf-8
        //得到session
        HttpSession session = req.getSession();
        //给session存东西
        session.setAttribute("name",new Person("秦江",1));
        //获取session的id
        String id = session.getId();
        //判断是否是新的Sessio
        if(session.isNew()){
            resp.getWriter().write("session创建成果,ID: "+id);
        }else{
            resp.getWriter().write("session已经存在");
        }
        //Session创建的时候做了什么事情,将它放到cookie里面,本质像cookie
//        Cookie cookie = new Cookie("JSESSIONID",id);
//        resp.addCookie(cookie);
    }

    @Override
    protected void doPost(HttpServletRequest req, HttpServletResponse resp) throws ServletException, IOException {
        doGet(req, resp);
    }
}
```
## 取得Session的值
```java
package com.kuang.servlet;

import com.kuang.pojo.Person;

import javax.servlet.ServletException;
import javax.servlet.http.HttpServlet;
import javax.servlet.http.HttpServletRequest;
import javax.servlet.http.HttpServletResponse;
import javax.servlet.http.HttpSession;
import java.io.IOException;

public class SessionDemo02 extends HttpServlet {
    @Override
    protected void doGet(HttpServletRequest req, HttpServletResponse resp) throws ServletException, IOException {
        req.setCharacterEncoding("UTF-8");
        resp.setCharacterEncoding("UTF-8");
        resp.setContentType("text/html;charset=utf-8");
        HttpSession session = req.getSession();
        Person person = (Person) session.getAttribute("name");//获取
        System.out.println(person.toString());

    }

    @Override
    protected void doPost(HttpServletRequest req, HttpServletResponse resp) throws ServletException, IOException {
        doGet(req, resp);
    }
}
```
## 删除Session
```java
package com.kuang.servlet;

import javax.servlet.ServletException;
import javax.servlet.http.HttpServlet;
import javax.servlet.http.HttpServletRequest;
import javax.servlet.http.HttpServletResponse;
import javax.servlet.http.HttpSession;
import java.io.IOException;

//注销session
public class SessionDemo03 extends HttpServlet {
    @Override
    protected void doGet(HttpServletRequest req, HttpServletResponse resp) throws ServletException, IOException {
        HttpSession session = req.getSession();
        session.removeAttribute("name");
        session.invalidate();//手动注销
    }

    @Override
    protected void doPost(HttpServletRequest req, HttpServletResponse resp) throws ServletException, IOException {
        doGet(req, resp);
    }
}
```
或者在web.xml中设置自动过期
```XML
<!--设置session默认消除-->
    <session-config>
<!--        15分钟后session自动失效-->
        <session-timeout>15</session-timeout>
    </session-config>
```
